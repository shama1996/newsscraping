```python
# Importing libraries.
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException,ElementNotInteractableException
```

```python
# Opening webdriver.
driver = webdriver.Chrome(r"C:\Users\HOMA\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
# Url of the website.
url = "https://timesofindia.indiatimes.com/archive/year-2019,month-12.cms"
```

```python
# Getting url.
driver.get(url)
```

```python
# Getting Urls of all days.
urls=driver.find_elements_by_xpath("//*[@id='calender']/tbody/tr/td/a")
UR=[]
for i in urls:
    UR.append(i.get_attribute('href'))
```

```python
# All days url.
UR
```

```python
# Scraping data of headline through exception handling for all the urls dates.
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]

for u in UR:
    driver.get(u)
    try:                  #exception handling for nosuchelementexception\n",
        head=driver.find_element_by_xpath("//table[@class='cnt']/tbody/tr[2]/td[1]/div[3]/table/tbody/tr[2]/td[1]/span/a")
        head.click()                                                      
        heading=driver.find_element_by_xpath("//h1[@class='_23498']")
        Headline.append(heading.text)
    except NoSuchElementException   as e:
        Headline.append("null")
```

```python
# Scraping data of author through exception handling for all the urls dates.
Author=[]
for u in UR:
    driver.get(u) 
    try:                  #exception handling for nosuchelementexception\n",
        head=driver.find_element_by_xpath("//table[@class='cnt']/tbody/tr[2]/td[1]/div[3]/table/tbody/tr[2]/td[1]/span/a")
        head.click()                                                      
        aut=driver.find_element_by_xpath("//a[@class='auth_detail']")
        Author.append(aut.text)
    except NoSuchElementException   as e:
        Author.append("null")
```

```python
# Checking author.
Author
```

```python
# Scraping data of Date through exception handling for all the urls dates.
Date=[]
for u in UR:
    driver.get(u)
    try:                  #exception handling for nosuchelementexception\n",
        head=driver.find_element_by_xpath("//table[@class='cnt']/tbody/tr[2]/td[1]/div[3]/table/tbody/tr[2]/td[1]/span/a")
        head.click()                                                      
        dat=driver.find_element_by_xpath("//div[@class='_3GNSx']")
        Date.append(dat.text)
    except NoSuchElementException   as e:
        Date.append("null")
```

```python
# Scraping data of Description through exception handling for all the urls dates.
Description=[]
for u in UR:
    driver.get(u) 
    try:                  #exception handling for nosuchelementexception\n",
        head=driver.find_element_by_xpath("//table[@class='cnt']/tbody/tr[2]/td[1]/div[3]/table/tbody/tr[2]/td[1]/span/a")
        head.click()                                                      
        des=driver.find_element_by_xpath("//*[@id='app']/div/div[3]/div[2]/div[1]/div[2]/div[3]/div[2]/div/text()[1]")
        Description.append(des.text)
    except NoSuchElementException   as e:
        Description.append("null")
```

```python
# Checking Date.
Date
```

```python
# Checking Description.
Description
```

```python
# Creating DataFrame.
News_classification =pd.DataFrame({"headline":Headline,
                "Date":Date,
                "Author":Author,
                "description":Description})
```

```python
# Final DataFrame.
News_classification
```

```python
# Converting to csv format.
News_classification.to_csv('News_classification.csv', index = False)
```
