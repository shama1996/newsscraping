```python
# Importing Libraries.
import selenium
from selenium import webdriver
import pandas as pd
```

```python
# Opening driver.
driver = webdriver.Chrome(r"C:\Users\HOMA\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
# URL of the website.
url = "https://timesofindia.indiatimes.com/archive/year-2019,month-12.cms"
```

```python
# Getting URL.
driver.get(url)
```

```python
# Finding URLS of the all dates through href.
urls=driver.find_elements_by_xpath("//*[@id='calender']/tbody/tr/td/a")
url=[]
for i in urls:
    url.append(i.get_attribute('href'))
```

```python
# Checking urls's.
url
```

```python
# Creating empty sets.
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
```

```python
#opening the first url of 1st december.
driver.get(url[0])
URLS1=[] #appending all the urls of the news from 1st december
for i in driver.find_elements_by_xpath("//div[@style='font-family:arial ;font-size:12;font-weight:bold; color: #006699']/table/tbody/tr[2]/td/span/a"):
    URLS1.append(i.get_attribute("href"))
```

```python
# Data of First date url in the month of december.
URLS1
```

```python
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
#News from 1st december will be appended here with their dates,headline,vertical,description and Author.
for i in URLS1[1:20]:
    driver.get(i)
    try:
        date = driver.find_element_by_xpath("//div[@class='_3lvqr clearfix']/div[2]/div/div").text
        Date.append(date)
    except NoSuchElementException:
        Date.append('-')
            
    try:
        des = driver.find_element_by_xpath("//div[@class='_1_Akb clearfix  ']/div").text
        Description.append(des)
    except NoSuchElementException:
        Description.append('-')
    
    try:
        headline = driver.find_element_by_xpath("//div[@class='_3lvqr clearfix']/div[2]/h1").text
        Headline.append(headline)
    except NoSuchElementException:
        Headline.append('-')
    
    try:
        vertical = driver.find_element_by_xpath("//li[@class='_8aEAx   K-f4t']/a").text
        Vertical.append(vertical)
    except NoSuchElementException:
        Vertical.append('-')
    
    try:
        author = driver.find_element_by_xpath("//div[@class='_3lvqr clearfix']/div[2]/div/div/span/a").text
        Author.append(author)
    except NoSuchElementException:
        Author.append('-')
```
